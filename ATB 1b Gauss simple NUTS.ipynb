```python
import pymc3 as pm
import matplotlib.pyplot as plt

# Create pyMC3 model object
gauss_model = pm.Model()

with gauss_model:
    x = pm.Normal('x', mu=3, sd=1)
    
    # Draw 500 posterior samples
    trace = pm.sample(500)
    
    pm.summary(trace)
    pm.traceplot(trace)
    
    plt.show()
```
